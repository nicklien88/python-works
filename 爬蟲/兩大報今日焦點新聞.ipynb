```python
import requests
from bs4 import BeautifulSoup

def get_web_page(url):
    resp = requests.get(
        url=url
    )
    if resp.status_code != 200:
        print('Invalid url:',resp.url)
        return None
    else:
        return resp.text
        
print("蘋果日報今日焦點")
dom = requests.get("https://tw.appledaily.com/appledaily/hotdaily/headline").text
soup = BeautifulSoup(dom, "html5lib")
for ele in soup.find("ul", "headline").find_all("li"):
    print(
        ele.find("div", "aht_title_num").text,
        ele.find("div", "aht_title").text
    )
print("----------------")
print("自由時報今日焦點")
dom = requests.get("https://news.ltn.com.tw/list/breakingnews").text
soup = BeautifulSoup(dom, "html5lib")
for ele in soup.find("ul", "list").find_all("li"):
    print(ele.find("a", "tit").p.text.strip())
```
