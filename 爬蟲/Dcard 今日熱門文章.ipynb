```python
import requests
from bs4 import BeautifulSoup
import re

URL = "https://www.dcard.tw/f"
resp = requests.get(URL)
soup = BeautifulSoup(resp.text, "html.parser")

articles = []

#利用 regex 找出所有 "PostList_wrapper" 開頭的 div
for  div in soup.find_all("div", re.compile("PostList_entry_\w{6}")):
    articles.append(
        {"Title":div.find("h3").text.strip(),
         "Excerpt":div.find("article").find_all("div")[2].text.strip(),
         "Bookmark":re.findall(r'\d+', div.find("article").find_all("div")[3].text.strip())[0],
         "Response":re.findall(r'\d+', div.find("article").find_all("div")[3].text.strip())[1],
         "Href":div.a["href"]
        })

print("共 %d 篇" % (len(articles)))
count = 0
for article in articles:
    print(count, article)
    count +=1
```
